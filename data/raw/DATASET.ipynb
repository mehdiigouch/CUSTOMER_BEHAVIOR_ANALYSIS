```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
PROJECT_ROOT = Path.cwd().parent.parent.parent.parent.parent
print(PROJECT_ROOT)
DATASET_PATH = PROJECT_ROOT / "PROJECTS" / "RETAIL COMPANY PROJECT" / "DATASET" / "customer_shopping_behavior.csv"
print(DATASET_PATH)
```

```python
dataset = pd.read_csv(DATASET_PATH)
df = dataset.copy()
df.head()
```

```python
df.info()
```

```python
df.columns=df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df.rename(columns={"purchase_amount_(usd)":"purchase_amount"},inplace=True)
df.columns
```

```python
df.describe(include="all").T
```

```python
df.isnull().sum()
```

```python
df["review_rating"].describe()
```

```python
df["review_rating"] = df.groupby("category")["review_rating"].transform(lambda x: x.fillna(x.median()))

df.isnull().sum()
```

```python
df["age"].describe()
```

```python
def age_category(age):
    if age < 18:
        return "Child"
    elif age < 30:
        return "Young Adult"
    elif age < 45:
        return "Adult"
    elif age < 60:
        return "Senior Adult"
    else:
        return "Elder"

df["age_group"] = df["age"].apply(age_category)

df[["age", "age_group"]].head(10)
```

```python
df["frequency_of_purchases"].unique()

```

```python
frequency_mapping = {

    "Weekly" : 7,
    "Fortnightly" : 14,
    "Biweekly" : 14,
    "Monthly" : 30,
    "Quarterly" : 90, 
    "Every 3 Months" : 90,
    "Annually" : 365
}


df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)


df[["frequency_of_purchases", "purchase_frequency_days"]].head(10)
```
